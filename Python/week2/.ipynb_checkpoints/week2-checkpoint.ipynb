```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.loc[df['Income.Group']=='<=50K','Income.Group'] = 0
df.loc[df['Income.Group']=='>50K','Income.Group'] = 1
df['Income.Group'] = df['Income.Group'].astype(int)
```

```python
df.head()
```

```python
df1 = pd.get_dummies(df)
```

```python
df1.head()
```

```python
features = list(set(df1.columns) - set(['ID','Income.Group']))
```

#### Principal Component Analysis

```python
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
```

```python
X = df1[features].values
```

```python
pca = PCA(n_components='mle')
```

```python
X_pca = pca.fit_transform(scale(X))
```

```python
X_pca
```

```python
pca.explained_variance_ratio_
```

```python
cum_var_exp = np.cumsum(np.round(pca.explained_variance_ratio_, decimals=4) * 100)
```

```python
cum_var_exp
```

```python
imp = pd.DataFrame(pca.components_,columns = features).T
```

```python
imp
```

# Modeling

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
X = df1[features]
Y = df1['Income.Group']
```

**Split the data to training and testing**

```python
trainX, testX, trainY, testY =  train_test_split(X, Y, test_size = .3, random_state = 166)
```

```python
dt = DecisionTreeClassifier()
```

```python
model = dt.fit(trainX,trainY)
```

```python
preds = model.predict(testX)
```

```python
accuracy = accuracy_score(testY, preds)
precision = precision_score(testY, preds)
recall = recall_score(testY, preds)
print accuracy,precision,recall
```

```python
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
```

### Hyper parameter Tuning

```python
from sklearn.grid_search import GridSearchCV
```

```python
parameters = {
    'criterion': ['gini', 'entropy'],
    'max_features': [None, 'auto', 'sqrt', 'log2'],
    'max_depth': [None, 1, 2, 5, 10],
    'min_samples_split': [1.0, 2, 5, 10]
}
```

```python
model = RandomForestClassifier(random_state=1)
```

```python
clf = GridSearchCV(model, parameters, cv=5, scoring='recall', verbose=True)
```

```python
clf.fit(trainX,trainY)
```

```python
print('Best score for data1:', clf.best_score_) 
```
